## EXCELを処理する関数

国土交通省ホームページのEXCELデータから作業する

輸出・輸入・移出・移入とも同じ構成のExcelになっているので、EXCELのインポートからDataFameの整形までは共通関数でOK

```python
import pandas as pd

def get_excel(url):
  # ブックの指定
  input_file_name = url
  input_book = pd.ExcelFile(input_file_name)
  
  # 統計結果は2番目のシート、上から2行は無視
  input_sheet_name = input_book.sheet_names
  df = input_book.parse(input_sheet_name[1],skiprows = 2)

  df = df.astype({'トン数': 'int'})

  return df
```

## 港別✕品種別

### 集計関数

```python
import pandas as pd
import numpy as np

def port_kind(url, params):

  # EXCELをインポート
  df = get_excel(url)

  # 集計
  df = pd.pivot_table(df, index=['都道府県', '港名'], columns='品種名', values='トン数', margins=True, aggfunc=np.sum).reset_index()

  # 都道府県指定がある場合
  if 'prefecture' in params:
    df = df[df['都道府県'].isin(params['prefecture'])]

  # 港湾指定がある場合
  if 'port' in params:
    df = df[df['港名'].isin(params['port'])]

  # 品種指定がある場合
  if 'kind' in params:
    df = df[['都道府県','港名',] + params['kind']]

  # 欠損列を削除
  df = df.dropna(how='all', axis=1)

  return df
```

```python
# 年次指定
time = '2020年'

# パラメータ設定
params = {
    'prefecture': ['兵庫県', '大阪府', '和歌山県', '岡山県','香川県'],
    'port':['姫路','東播磨','水島','坂出'],
    'kind':['ＬＮＧ（液化天然ガス）','重油','原油','その他の石油','石炭'],

}

df = port_kind(get_url(time,'輸入'), params)
df
```

### Excelに保存

輸入、輸出、移入、移出、搬入（輸入＋移入）、搬出（輸出＋移出）、取扱貨物量（全合計）をExcelシートに保存

```python
def save_xls_port_kind(time,params):
  # 輸入
  df_import = port_kind(get_url(time,'輸入'), params)
  
  # 輸出
  df_export = port_kind(get_url(time,'輸出'), params)

  # 移入
  df_immigration = port_kind(get_url(time,'移入'), params)

  # 移出
  df_emigration = port_kind(get_url(time,'移出'), params)

  # 搬入（輸入＋移入）
  df_carryin = pd.concat([df_import, df_immigration]).groupby(['都道府県','港名'], as_index=False).sum()

  # 搬出（輸出＋移出）
  df_carryout = pd.concat([df_export, df_emigration]).groupby(['都道府県','港名'], as_index=False).sum()

  # 取扱貨物量（総計）
  df_cargo = pd.concat([df_carryin, df_carryout]).groupby(['都道府県','港名'], as_index=False).sum()
  
  with pd.ExcelWriter(f'港湾統計_港名✕品種別【{time}】.xlsx') as writer:
    df_import.to_excel(writer, sheet_name='輸入')
    df_export.to_excel(writer, sheet_name='輸出')
    df_immigration.to_excel(writer, sheet_name='移入')
    df_emigration.to_excel(writer, sheet_name='移出')
    df_carryin.to_excel(writer, sheet_name='搬入（輸入＋移入）')
    df_carryout.to_excel(writer, sheet_name='搬出（輸出＋移出）')
    df_cargo.to_excel(writer, sheet_name='取扱貨物量（総計）')
    
  files.download(f'港湾統計_港名✕品種別【{time}】.xlsx')

  return
```

```python
# 年次指定
time = '2020年'

# パラメータ設定
params = {
    'prefecture': ['兵庫県', '大阪府', '和歌山県', '岡山県','香川県'],
    'port':['姫路','東播磨','水島','坂出'],
    'kind':['ＬＮＧ（液化天然ガス）','重油','原油','その他の石油','石炭'],
}

# Excelに保存
save_xls_port_kind(time,params)
```

## 港別✕仕向国別（品種指定）

```python
import pandas as pd
import numpy as np

def import_country(url,kind):

  # EXCELをインポート
  df = get_excel(url)

  # 品種名（kind）に一致する行を抽出
  df = df[df['品種名'] == kind]

  # 集計
  df = pd.pivot_table(df, index=['都道府県', '港名'], columns='仕向国', values='トン数', margins=True, aggfunc=np.sum).reset_index()

  return df
```

```python
# URLを指定
url = 'https://www.mlit.go.jp/k-toukei/content/001448038.xlsx'

# 品種名を指定
kind = 'ＬＮＧ（液化天然ガス）'

df = import_kind(url,kind)
df
```
